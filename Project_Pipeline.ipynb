### 1.3. Scraping Function
#### 1.3.1. Version 1

```python
def scrape_listV1(url_ex, url_men, export_df, mentions_df):
    global j
    '''
    This function will use the list of export.csv and mentions.csv files to cash their contents and only keep relavant
    columns
    '''
    for i in range(url_ex.shape[0]):
        # Appending is slightly faster than Concat when ignore_index=True, so we used append to add  new scraped dataFrame
        ## But appending gets inefficient for large dataFrame, so instead of appending the new scraped dataframe to a ...
        ## ... large dataFrame, we recursively call our function to use a new empty dataFrame for appending to achieve...
        ## ... much faster speed in scraping large number of dataframes
        if i>= 100:
            export_df_2 = pd.DataFrame(columns=col_ex_list)
            mentions_df_2 = pd.DataFrame(columns=col_men_list)
            export_df_2, mentions_df_2 = scrape_listV1(url_ex.iloc[100:], url_men.loc[100:], export_df_2, mentions_df_2)
            export_df = pd.concat([export_df, export_df_2], ignore_index=True, sort=False, copy=False)
            mentions_df = pd.concat([mentions_df, mentions_df_2], ignore_index=True, sort=False, copy=False)
            break
        else:
            try:
                df_i_x=pd.read_csv(url_ex.iloc[i], sep='\t', compression='zip', names=col_ex)
                df_i_m=pd.read_csv(url_men.iloc[i], sep='\t',compression='zip', names=col_men)
                export_df = pd.concat([export_df, df_i_x[col_ex_list]], ignore_index=True, sort=False, copy=False)
                mentions_df = pd.concat([mentions_df, df_i_m[col_men_list]], ignore_index=True, sort=False, copy=False)
            except urllib.error.HTTPError:
                pass
    return export_df, mentions_df
     
```

```python
def scrape_listV2(url_ex, url_men, export_df, mentions_df):
    global j
    '''
    This function will use the list of export.csv and mentions.csv files to cash their contents and only keep relavant
    columns
    '''
    for i in range(url_ex.shape[0]):
        # We use concatenating to achieve higher speed here...
        ## But for concatenating large dataFrames, we recursively call our function to use...
        ## ... a new empty dataFrame for appending to achieve much faster speed in scraping large number of dataframes
        if i>= 100:
            export_df_2 = pd.DataFrame(columns=col_ex_list)
            mentions_df_2 = pd.DataFrame(columns=col_men_list)
            export_df_2, mentions_df_2 = scrape_listV2(url_ex.iloc[100:], url_men.loc[100:], export_df_2, mentions_df_2)
            export_df = pd.concat([export_df, export_df_2], ignore_index=True, sort=False, copy=False)
            mentions_df = pd.concat([mentions_df, mentions_df_2], ignore_index=True, sort=False, copy=False) 
            break
        else:
            s_ex=requests.get(url_ex.iloc[i])
            s_men = requests.get(url_men.iloc[i])
            if s_ex.status_code==200 and s_men.status_code==200:
                df_i_x=pd.read_csv(io.BytesIO(s_ex.content), sep='\t', compression='zip', names=col_ex, low_memory=False)
                df_i_m=pd.read_csv(io.BytesIO(s_men.content), sep='\t',compression='zip', names=col_men, low_memory=False)
                export_df = pd.concat([export_df, df_i_x[col_ex_list]], ignore_index=True, sort=False, copy=False)
                mentions_df = pd.concat([mentions_df, df_i_m[col_men_list]], ignore_index=True, sort=False, copy=False)
    return export_df, mentions_df
     
```

## 4. Adding 'SourceCountry' Feature

### 4.1. Functions for mapping 'MentionSourceName' to their 'SourceCountry'

```python

def get_map_site():
    file = pd.read_csv(COUNTRY_CODE_DATA)
    return dict(zip(file['TLD'], file['ISO3166-1-Alpha-3'])), dict(zip(file['ISO3166-1-Alpha-3'], file['TLD']))

def src_to_country_v2(web_data): # Look at extension if fail go for ip alpha 3 
    def extension_lookup(website):
        try:
            return site[str('.') + website.split('.')[-1]]
        except:
            return None
        
    def ip_lookup(website):
        try:
            requests.get(website).status_code==200
            #s_re.status_code==200
            ip = socket.gethostbyname(website)
            #print('IP: ',ip)
            reader = geolite2.reader()
            output = reader.get(ip)
            #print('Output: ', output)
            results = output['country']['iso_code']
            #print('Results: ', results)
            country = pycountry.countries.get(alpha_2=results)
            return country.alpha_3
        except:
            return np.nan
        
    def two_way_lookup(website):
        ret = extension_lookup(website)
        if ret == None:
            ret = ip_lookup(website)
            return ret
        return ret
    
    web_list = []
    for x in web_data:
        y=two_way_lookup(x)
        web_list.append(y)

    return web_list
```

## 5.Updating the 'df_sourceName' dataframe

### 5.1. The function to update 'df_SourceName' dataframe

```python
def update_sourceName(source_df):
    def extension_lookup(website):
        try:
            return site[str('.') + website.split('.')[-1]]
        except:
            return None
        
    def ip_lookup(website):
        socket.setdefaulttimeout(1.5)
        
        try:
            ip = socket.gethostbyname(website)
            #print('This is IP: ', ip)
            #socket.close()
            #print('IP: ',ip)
            reader = geolite2.reader()
            output = reader.get(ip)
            #print('Output: ', output)
            results = output['country']['iso_code']
            #print('Results: ', results)
            country = pycountry.countries.get(alpha_2=results)
            return country.alpha_3
        except:
            return np.nan
        
    def two_way_lookup(x):
        #print('**************')
        #print('This is x: ',x)
        ret = extension_lookup(x)
        if ret == None:
            ret = ip_lookup(x)
            #print('This is ret: ', ret)
            return ret
        #print('This is ret: ', ret)
        return ret
    
    source_df = source_df.copy()
    df_dict = source_df[source_df.iloc[:,1].isna()].copy()
    df_dict['SourceCountry'] = df_dict.apply(lambda x: two_way_lookup(x.iloc[0]), axis=1)
    source_df[source_df.iloc[:,1].isna()] = df_dict #source_df.apply(lambda x: x.fillna(two_way_lookup(x.iloc[0])), axis=1)
    return source_df
```
